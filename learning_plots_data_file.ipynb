```python
daily_returns = np.log(df4).diff().dropna()
daily_returns.head()
```

```python
import plotly.graph_objects as go

# Check what columns are available
print("Available columns:", daily_returns.columns.tolist())

fig = go.Figure()

# Add trace for each available column that contains AAPL or GOOG
available_stocks = []
for col in daily_returns.columns:
    if 'AAPL' in col or 'GOOG' in col or 'GOOGL' in col:
        available_stocks.append(col)

print("Found stocks:", available_stocks)

# Plot the available stocks
for column in available_stocks:
    fig.add_trace(go.Scatter(
        x=daily_returns.index,
        y=daily_returns[column],
        mode='lines',
        name=column,
        line=dict(width=1)
    ))

fig.update_layout(
    title='Daily Log Returns',
    xaxis_title='Date',
    yaxis_title='Daily Log Returns',
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

```python
# Calculate annualized returns
annualized_returns = daily_returns.mean() * 252

fig = go.Figure()

fig.add_trace(go.Bar(
    x=annualized_returns.index,
    y=annualized_returns.values,
    name='Annualized Returns'
))

fig.update_layout(
    title='Annualized Returns',
    xaxis_title='Stocks',
    yaxis_title='Annualized Return',
    template='plotly_white'
)

fig.show()
```

```python
# To calculate 5 days rolling returns, simply sum daily returns for 5 days as log returns are additive
rolling_return = daily_returns.rolling(5).sum().dropna()
# Display the last five rows of the data frame to check the output
rolling_return.head(5)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=rolling_return.index,
    y=rolling_return['NFLX'],
    mode='lines',
    name='NFLX',
    line=dict(color='red', width=2)
))

fig.update_layout(
    title='5-Days Rolling Returns of NetFlix',
    xaxis_title='Date',
    yaxis_title='5-Day Rolling Return',
    template='plotly_white',
    hovermode='x'
)

fig.show()
```
